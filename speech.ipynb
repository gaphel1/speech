```python
import os
import librosa
import librosa.display
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import IPython.display
from tensorflow.python.platform import gfile
```

```python
PATH_TRAIN =r'C:\Users\Turing Gaphel\Documents\3new\dataset\train'
```

```python
path= os.path.join(PATH_TRAIN, '*','*.wav')
wave_path= gfile.Glob(path)[-1]
```

```python
IPython.display.Audio(wave_path)
```

```python
wave, sr= librosa.load(wave_path,mono=True)
plt.figure(figsize=(12,4))
librosa.display.waveplot(wave,sr=sr)
```

```python
mfccs=librosa.feature.mfcc(y=wave,sr=sr,n_mfcc=20)
plt.Figure(figsize=(12,4))
plt.title('MFCC')
librosa.display.specshow(mfccs,x_axis='time')
plt.colorbar()
```

```python
LABEL_TO_INDEX_MAP={'cat':0,'tree':1,'bed':2,'dog':3,'bird':4,'house':5}
_,label = os.path.split(os.path.dirname(wave_path))
encoding =[0] *len(LABEL_TO_INDEX_MAP)
encoding[LABEL_TO_INDEX_MAP[label]]=1
print('Wave path',wave_path)
print('Eabel',label)
print('Encoding:',encoding)
```
